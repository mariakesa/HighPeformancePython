This notebook implements the examples from the LAFF on Programming for High Performance in Python http://www.cs.utexas.edu/users/flame/laff/pfhp/section-1.html.

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def myGEMM(m,n,k):
    A=np.random.randn(m,k)
    B=np.random.randn(k,n)
    C=np.random.randn(m,n)
    start=time.time()
    for i in range(0,m):
        for j in range(0,n):
            for p in range(0,k):
                C[i,j]=A[i,p]*B[p,j]
    end=time.time()
    return end-start

times=[]
flops=[]
for j in range(1,500,100):
    m=n=k
    tim=myGEMM(m,n,k)
    times.append(tim)
    flops.append((2*m*n*k/tim)*10**-9)
```

```python
plt.plot(np.arange(1,500,50),times)
plt.xlabel('matrix dimension m=n=k')
plt.ylabel('time, seconds')
plt.title('IJP')
plt.show()


plt.plot(np.arange(1,500,50),flops)
plt.xlabel('matrix dimension m=n=k')
plt.ylabel('time, seconds')
plt.title('GFLOPS')
plt.show()
```
